## Merging DataFrame

```python
station_info = pd.read_csv('data/weather_stations.csv')
station_info.head()
```

### inner join

```python
inner_join = weather.merge(station_info,left_on = 'station',right_on='id')
inner_join.sample(5,random_state=0)
```

### id station   

```python
weather.merge(station_info.rename(dict(id='station'),axis=1),on='station').sample(5,random_state=0)
```

### Left Join

```python
left_join = station_info.merge(weather,left_on = 'id',right_on = 'station',how='left')
```

```python
left_join.head()
```

### right join

```python
right_join = weather.merge(station_info,left_on='station',right_on = 'id',how='right')
right_join.head()
```

```python
right_join[right_join.datatype.isna()].head()
```

### Full outer join

```python
outer_join = weather.merge(station_info[station_info.id.str.contains('US1NY')],left_on='station',right_on='id',
                           how='outer',indicator=True)
```

```python
outer_join.head()
```

```python
# View effect of outer join
pd.concat([outer_join.query(f'_merge == "{kind}"').sample(2,random_state=0) for kind in outer_join._merge.unique()]).sort_index()
```

### Window calculations

```python
central_park_weather.loc['2018-10'].assign(rolling_PRCP=lambda x: x.PRCP.rolling('3D').sum())[['PRCP','rolling_PRCP']].head(7).T
```

```python
central_park_weather.loc['2018-10'].rolling('3D').mean().head(7).iloc[:,:6]
```

### To apply different aggregations across columns

```python
central_park_weather['2018-10-01':'2018-10-07'].rolling('3D').agg({'TMAX':'max','TMIN':'min',
                                                                   'AWND':'mean','PRCP':'sum'}).join(central_park_weather[['TMAX','TMIN','AWND','PRCP']],lsuffix='_rolling').sort_index(axis=1)
```

## expanding windows

```python
central_park_weather.loc['2018-06'].assign(TOTAL_PRCP = lambda x: x.PRCP.cumsum(),AVG_PRCP=lambda x:x.PRCP.expanding().mean()).head(10)[['PRCP','TOTAL_PRCP','AVG_PRCP']].T
```

### Provide column-specific aggregations

```python
central_park_weather['2018-10-01':'2018-10-07'].expanding().agg({'TMAX':np.max,'TMIN':np.min,
                                                                 'AWND':np.mean,'PRCP':np.sum}).join(central_park_weather[['TMAX','TMIN','AWND','PRCP']],lsuffix='_expanding').sort_index(axis=1)
```

### Exponentially weighted moving windows

```python
central_park_weather.assign(
    AVG = lambda x: x.TMAX.rolling('30D').mean(),
    EWMA = lambda x: x.TMAX.ewm(span=30).mean()
).loc['2018-09-29':'2018-10-08',['TMAX','EWMA','AVG']].T
```
