```python
weather = pd.read_csv('weather_by_station.csv', index_col='date', parse_dates=True)
weather.head()
```

```python
weather[weather.datatype == "PRCP"].groupby(level=0).mean().groupby(pd.Grouper(freq='Q')).sum().value.nlargest()
```

```python
import numpy as np
```

```python
weather.query('datatype == "PRCP"').rename(dict(value='prcp'), axis=1).groupby(level=0).mean()
```

```python
weather.query('datatype == "PRCP"').rename(dict(value='prcp'), axis=1).groupby(level=0).mean().assign(total_prcp_in_quarter=lambda x:x.prcp.groupby(pd.Grouper(freq='Q')).transform(np.sum),
        numb_days_in_month = lambda x: x.prcp.groupby(pd.Grouper(freq='M')).transform('count') )
```

```python
weather.query('datatype == "PRCP"').rename(dict(value='prcp'), axis=1).groupby(level=0).mean().assign(
        total_prcp_in_quarter=lambda x:
            x.prcp.groupby(pd.Grouper(freq='Q')).transform(np.sum),
        numb_days_in_month = lambda x: (x.prcp == x.prcp).groupby(pd.Grouper(freq='M')).transform(np.sum) )
```

```python
fb = pd.read_csv('fb_2018.csv', index_col='date', parse_dates=True).assign(
    trading_volume=lambda x: pd.cut(x.volume, bins=3, labels=['low', 'med', 'high'])
)
fb.head()
```

```python
pd.crosstab(
    index=fb.trading_volume,
    columns=fb.index.month,
    colnames=['month']
)
```

```python
fb.pivot_table(
    index='trading_volume',
    columns=fb.index.month,
    values = 'volume',
    aggfunc="count")
```
