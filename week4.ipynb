```python
df.date = pd.to_datetime(df.date)
```

```python
station_qm_wesf = df[df.station=='?'].drop_duplicates('date').set_index('date').WESF
```

```python
df.sort_values('station',ascending=False,inplace=True)
```

```python
df_deduped = df.drop_duplicates('date')
```

```python
df_deduped = df_deduped.drop(columns='station').set_index('date').sort_index()
```

```python
df_deduped = df_deduped.assign(WESF = lambda x:x.WESF.combine_first(station_qm_wesf))
```

```python
df_deduped.shape
```

```python
df_deduped.dropna().shape
```

```python
df_deduped.dropna(how="all").shape
```

```python
df_deduped.dropna(how='all',subset=['inclement_weather','SNOW','SNWD']).shape
```

```python
df_deduped.dropna(axis='columns',thresh=df_deduped.shape[0] *.75).columns
```

```python
df_deduped.loc[:,'WESF'].fillna(0,inplace=True)
df_deduped.head()
```

### Replace unreasonable values with NaN

```python
df_deduped = df_deduped.assign(TMAX = lambda x: x.TMAX.replace(5505,np.nan),
                               TMIN = lambda x: x.TMIN.replace(-40,np.nan))
```

```python
df_deduped.assign(TMAX = lambda x: x.TMAX.fillna(method="ffill"),
                  TMIN = lambda x: x.TMIN.fillna(method="ffill")).head()
```

```python
df_deduped.assign(SNWD = lambda x: np.nan_to_num(x.SNWD)).head()
```

### The clip method makes it possible to cap values at a specific minimum and/or maximum threshold Snow depth can't be negative

```python
df_deduped.assign(SNWD = lambda x:x.SNWD.clip(0,x.SNOW)).head()
```

```python
df_deduped.assign(
    TMAX = lambda x: x.TMAX.fillna(x.TMAX.median()),
    TMIN = lambda x: x.TMIN.fillna(x.TMIN.median()),
    TOBS = lambda x: x.TOBS.fillna((x.TMAX+x.TMIN)/2)
)
```

### Rolling 7-day median

```python
df_deduped.apply(lambda x: x.fillna(x.rolling(7,min_periods=0).median())).head(10)
```

### linear interpolation

```python
df_deduped.reindex(
    pd.date_range('2018-01-01','2018-12-31',freq='D')
).apply(lambda x: x.interpolate()).head(10)
```
