### pivoting and unstack

```python
long.columns
```

```python
long_df = pd.read_csv(
    'long_data.csv', usecols=['date', 'datatype', 'value']
).rename(
    columns={'value': 'temp_C'}
).assign(
    date=lambda x: pd.to_datetime(x.date),
    temp_F=lambda x: (x.temp_C * 9/5) + 32
)
```

```python
long_df.head(10)
```

date를 인덱스로 고정 하고 long_df의 칼럼 중 'datatype'의 값을 가져온다

```python
pivoted_df = long_df.pivot(index='date',columns='datatype',values='temp_C')
pivoted_df.head(10)
```

```python
pivoted_df = long_df.pivot(index='date',columns='datatype',values=['temp_C','temp_F'])
pivoted_df.head()
```

### Multi index를 이용해 'date'와 'datatype'을 index로 구성

```python
multi_index_df = long_df.set_index(['date','datatype'])
multi_index_df.head().index
```

```python
multi_index_df.head()
```

```python
unstacked_df = multi_index_df.unstack()
unstacked_df.head()
```

```python
unstacked_df = pivoted_df.unstack()
unstacked_df
```

```python
station_qm_wesf = df[df.station=='?'].drop_duplicates('date').set_index('date').WESF
```

```python
df.sort_values('station',ascending=False,inplace=True)
```

```python
df_deduped = df.drop_duplicates('date')
```

```python
df_deduped = df_deduped.drop(columns='station').set_index('date').sort_index()
```

```python
df_deduped = df_deduped.assign(WESF = lambda x:x.WESF.combine_first(station_qm_wesf))
```

```python
df_deduped.dropna(how="all").shape
```

```python
df_deduped.dropna(how='all',subset=['inclement_weather','SNOW','SNWD']).shape
```

```python
df_deduped.dropna(axis='columns',thresh=df_deduped.shape[0] *.75).columns
```

```python
df_deduped.loc[:,'WESF'].fillna(0,inplace=True)
df_deduped.head()
```

```python
df_deduped = df_deduped.assign(TMAX = lambda x: x.TMAX.replace(5505,np.nan),
                               TMIN = lambda x: x.TMIN.replace(-40,np.nan))
```

```python
df_deduped.assign(TMAX = lambda x: x.TMAX.fillna(method="ffill"),
                  TMIN = lambda x: x.TMIN.fillna(method="ffill")).head()
```
